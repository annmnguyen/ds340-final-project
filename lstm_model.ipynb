```python
import os
import requests
import pandas as pd



def download_audio(url, file_name):
    # Create a directory for downloaded audios if it doesn't exist
    directory = 'downloaded_audios'
    os.makedirs(directory, exist_ok=True)

    # Define the path where the file will be saved
    file_path = os.path.join(directory, sanitize_filename(file_name) + '.mp3')

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raises a HTTPError for bad responses
        with open(file_path, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded successfully: {file_name}")
    except requests.RequestException as e:
        print(f"Failed to download {file_name}: {e}")
    
    return file_path

def sanitize_filename(name):
    # Clean and create a valid filename
    return "".join(char for char in name if char.isalnum() or char in " -_").rstrip()

# Iterate over the DataFrame rows
for _, row in total.iterrows():
    # Create a descriptive filename using track_id and track_name
    filename = f"{row['track_id']}_{row['track_name']}"
    download_audio(row['url'], filename)
```

```python
import librosa
import numpy as np

def extract_features(file_path, sample_rate=22050, num_mfcc=13, n_fft=2048, hop_length=512):
    # Load the audio file
    signal, sr = librosa.load(file_path, sr=sample_rate)
    # Extracting MFCCs
    mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length)
    return np.mean(mfcc.T, axis=0)  # Taking mean to simplify the feature array
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Example features and labels lists
features = []  # This should be filled with the MFCC data you've extracted
labels = []  # This should be filled with the corresponding genres or labels

# Convert lists to Numpy arrays
X = np.array(features)
y = np.array(labels)

# Encode categorical labels to integers
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=42)

# If you also need a validation set, you can split the test set further:
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

def build_model(input_shape):
    model = Sequential([
        LSTM(64, input_shape=input_shape, return_sequences=True),
        Dropout(0.5),
        LSTM(64),
        Dense(64, activation='relu'),
        Dense(len(label_encoder.classes_), activation='softmax')
    ])
    return model

# Prepare model for training
model = build_model((X_train.shape[1], X_train.shape[2]))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Model accuracy: {accuracy*100:.2f}%")

# Predict new samples
predictions = model.predict(X_test)
```
